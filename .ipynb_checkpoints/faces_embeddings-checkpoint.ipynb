```python
from skimage import io
import matplotlib.pyplot as plt
import face_embeddings_demo

img_path = "img.jpg"
image = io.imread(img_path)
```

```python
plt.imshow(image)
```

```python
face_rects = face_embeddings_demo.getFace(image)
```

```python
if len(face_rects) == 0:
    print("There aren't faces")
else:
    faces = []
    embeddings = []
    for rect in face_rects:
        faces += [rect['face']]
        embeddings += [face_embeddings_demo.getEmbedding(rect['face'])]
```

```python
num_columns = 2
for i, face, emb in zip(range(len(faces)), faces, embeddings):
    plt.figure(figsize=(15,15))
    plt_idx = i * num_columns + 1
    plt.subplot(len(faces), num_columns, plt_idx)
    plt.imshow(face)
    plt.axis('off')
    plt.figure(figsize=(50,30))
    plt_idx = i*num_columns + 2
    plt.subplot(len(faces), num_columns, plt_idx)
    plt.imshow(emb)
    plt.axis('off')
plt.show()
```
